```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
diamonds = pd.read_csv("/content/diamonds.csv", index_col=0)
```

```python
diamonds.info()
```

**Histogram**

```python
# histogram of carat
plt.hist(diamonds['carat'], bins=20, edgecolor='k',color='orange')
plt.title('Histogram of Carat')
plt.show()
```

**Bar Plot**

```python
plt.bar(diamonds['cut'], diamonds['price'])
plt.xlabel('Cut')
plt.ylabel('Price')
plt.title('Bar Plot of Color by Clarity')
plt.xticks(rotation=45)
plt.show()
```

**KDE Hist Plot of Price**

```python
##sns.kdeplot(data=diamonds, x="price", bw_adjust=.2)
sns.kdeplot(diamonds['price'], fill=True , color='r')
```

**Joint Plot of price versus carat**

```python
sns.jointplot(x='price', y='carat', data=diamonds)
```

**Heat map** **and correlation matrix**

```python
Var_Corr = diamonds.corr()
# plot the heatmap and annotation on it
sns.heatmap(Var_Corr, xticklabels=Var_Corr.columns, yticklabels=Var_Corr.columns, annot=True)
```
