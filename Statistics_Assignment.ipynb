# Case Study on Testing of Hypothesis

```python
sales_add = pd.read_csv('/content/Sales_add.csv')
sales_add
```

##a. The company wishes to clarify whether there is any increase in sales after stepping into digital marketing

```python
import scipy.stats as stats
from statsmodels.stats.weightstats import ztest as ztest
z_stat,p_val=ztest(sales_add['Sales_before_digital_add(in $)'],sales_add['Sales_After_digital_add(in $)'],value=0)
```

```python
p_val
```

```python
if p_val<0.05:
  print("Reject Null Hypothesis")
else:
   print("Accept Null Hypothesis")
```

##b. The company needs to check whether there is any dependency between the features “Region” and “Manager”.

```python
compare = pd.crosstab(sales_add['Region'],sales_add['Manager'])
compare
```

```python
chi2, p, dof, ex = stats.chi2_contingency(compare)
print(f'Chi_square value {chi2}\n\np value {p}\n\ndegrees of freedom {dof}\n\n expected {ex}')
```

```python
if p<0.05:
  print("Reject Null Hypothesis")
else:
   print("Accept Null Hypothesis")
```
