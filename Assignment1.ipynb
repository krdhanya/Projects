**Data visualizations using the dataset**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

***Scatter Plot***

```python
iris.plot(kind='scatter', x='SL', y='SW')
```

***Box Plot***

```python
sns.boxplot(x='Classification', y='PW', data=iris)
```

***Strip Plot***

```python
ax=sns.stripplot(x='Classification', y='PL', data=iris, jitter=True, edgecolor='gray')
```

***Violin Plot***

```python
sns.violinplot(x='Classification', y='PL', data=iris,size=6)
```

***Pair plot***

```python
sns.pairplot(iris,hue='Classification', height=3)
```
